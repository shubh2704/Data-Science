**6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.**

```python
class median:

#defining a constructor
  def __init__(self, numbers):
    self.numbers = numbers

#defining a median function to calculated median
  def calculate_median(self):
    sorted_numbers = sorted(self.numbers) #sort the dataset
    length = len(sorted_numbers) #to find the length of dataset

#if the length is even
    if length % 2 == 0:
      middle = int(length / 2)
      return (sorted_numbers[middle - 1] + sorted_numbers[middle])/2
#if the length is odd
    else:
      middle = int(length / 2)
      return (sorted_numbers[middle])
```

```python
#examples to show the constructor is working correctly
x = median([1,5,3,2])
y = median([1,5,3])
median1 = x.calculate_median()
median2 = y.calculate_median()
```

**10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.**

```python
#approach 1 by using basic python program where blank list total_bedroom_size is used
total_bedroom_size = []
for i in df['total_bedrooms']:
  if i <= 10: #if condition to quote as small
    total_bedroom_size.append('Small')
  elif i >= 11 and i <= 1000: #elif condition to quote as medium
    total_bedroom_size.append('Medium')
  else: #else to quote as large
    total_bedroom_size.append('Large')

df_new = df.copy()
df_new['total_bedroom_size'] = total_bedroom_size # to add new column total_bedroom_size

df_new.head()
```

```python
#Approach 2 using pandas loc function
new_df = df.copy()

# set the default value as large
new_df["total_bedroom_size"] = "Large"

# set values to "small" and "medium" based on the specified criteria
new_df.loc[new_df["total_bedrooms"] <= 10, "total_bedroom_size"] = "Small"
new_df.loc[(new_df["total_bedrooms"] > 10) & (new_df["total_bedrooms"] < 1000), "total_bedroom_size"] = "Medium"


new_df.head()
```
